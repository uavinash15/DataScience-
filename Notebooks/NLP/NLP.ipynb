```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
import spacy
import wordcloud
nlp= spacy.load('en_core_web_sm')
import nltk
nltk.download('all')
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
import string
```

```python
#### Stopwords
stop= stopwords.words('english')
print(stop)
```

```python
len(stop)
```

```python
#### punctuation
string.punctuation
```

```python
#### Tokenization
sent='Data Science class started in the month of November 2025. The class is going to complete on January 2026'
```

```python
print(word_tokenize(sent))
```

```python
print(sent_tokenize(sent))
```

```python
#### Lemmatization
lemmatizer= WordNetLemmatizer()
lemmatizer.lemmatize('running',pos='v')
```

```python
lemmatizer.lemmatize('taj mahal',pos='a')
```

```python
lemmatizer.lemmatize('books',pos='n')
```

```python
##### Stemming
ps= PorterStemmer()
ps.stem('beauty')
```

```python
ps.stem('taj mahal')
```

```python
doc=nlp(sent)
sent
```

```python
doc
```

```python
for token in doc:
  print(token.text,token.pos_,spacy.explain(token.pos_),
        token.lemma_,token.is_stop,token.is_punct,token.is_digit,token.is_bracket,token.is_currency)
```

#### Named Entity Recognition(NER)

```python
doc=nlp(sent)
doc
```

```python
for token in doc.ents:
  print(token.text,token.label_,spacy.explain(token.label_),token.start_char,token.end_char)
```

```python
spacy.displacy.render(doc,style='ent')
```

WordCloud

```python
courses= 'DataScience, DataAnalyst, GenAI, The AgentAI, BuisnessAnalyst, DataScience, GenAI, GenAI, JAVA, Python'
```

```python
from wordcloud import WordCloud ,STOPWORDS
```

```python
print(set(STOPWORDS))
```

```python
len(STOPWORDS)
```

```python
stop= STOPWORDS.update(['JAVA','Python'])
```

```python
word= WordCloud(width=1000,height=1000,background_color='white',stopwords=stop).generate(courses)
```

```python
word
```

```python
def word_cl(image):
  plt.figure(figsize=(10,10))
  plt.imshow(image)
  plt.axis('off')
  plt.savefig('wordcloud.png',format='png')
```

```python
word_cl(word)
```
