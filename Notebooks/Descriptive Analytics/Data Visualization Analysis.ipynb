```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#to upload data file
df1=pd.read_csv('Universities.csv')
```

```python
df1
```

```python
# to make histogram of 'SAT' column in df1 data set
plt.hist(df1['SAT'])
```

```python
# for not wanting the data (array([1., 3., 0., 0., 1., 2., 7., 4., 2., 5.]),
 #array([1005., 1046., 1087., 1128., 1169., 1210., 1251., 1292., 1333.,
        #1374., 1415.]), from above result give this code

plt.hist(df1['SAT'])
plt.show()
```

```python
plt.hist(df1['SAT'])
#to give x axis name
plt.xlabel('SAT SCORES')
#to give y axis name
plt.ylabel('Frequency')
#to give title
plt.title('Histogram of SAT scores')
# to just show the chart without the other intervals arrays
plt.show()
```

```python
#histplot is under seaborn library
sns.histplot(df1['SAT'])
plt.show()
```

```python
# to create a scatter plot
plt.plot(df1['SAT'],df1['Accept'])
```

```python
#to give only blue points in the scatter plot give 'bo' which gives blue points
plt.plot(df1['SAT'],df1['Accept'],'bo')
```

```python
#xlabel to give x axis name
#ylable to give y axis name
# title to give title of the scatter plot
#rs for red circles
plt.plot(df1['SAT'],df1['Accept'],'rs')
plt.xlabel('SAT Scores')
plt.ylabel('Acceptances')
plt.title('Scatter plot of SAT vs Acceptance')
plt.show()
```

```python
plt.scatter(df1['SAT'], df1['Top10'])
plt.xlabel('SAT Score')
plt.ylabel('Top 10%')
plt.title('SAT Score vs. Top 10%')
plt.show()
```
