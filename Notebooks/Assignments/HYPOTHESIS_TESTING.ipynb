# **HYPOTHESIS TESTING**

**1.Hypothesis Statement:**

```python
# H0(Null Hypothesis)= The model is still correct.
# H1(Alternate Hypothesis)= The real mean cost is higher than predicted
```

**2.Test Statistic**

```python
n=25 #sample size
print('sample size: ',n)
xbar=3050 #sample mean
print('sample mean: ',xbar)
X=600 # no of units produced in week
print('no of units produced in week',X)
std_X=25 #standard deviation of X
print('standard deviation of X: ',std_X)
sigma=5*std_X # population std
print('population standard deviation: ',sigma)
W=1000+5*X # Model
mu=W #Theoretical Mean
print('Theoretical Mean: ',W)
```

standard error(SE)

```python
import math
```

```python
SE=sigma/math.sqrt(n)
print('Standard Error: ',SE)
```

```python
Z=( xbar-mu )/SE
print('Z test statistic value: ',Z)
```

**3.Critical value**

```python
# right tailed test
alpha=0.05
print('significance level: ',alpha)
```

```python
import scipy.stats as stats
Z_critic=stats.norm.ppf(1-alpha)
print('Z critical value: ',Z_critic)
```

**4.Decision:**

```python
if(Z>Z_critic):
  print('Rejecting Null Hypothesis H0')
else:
  print('Failed to Reject Null Hypothesis H0')
```

**5.Conclusion**



Based on the hypothesis test, we failed to reject the null hypothesis ($H_0$). This means there is not enough statistical evidence at the 5% significance level to support the restaurant owners' claim that the real mean weekly operating cost is higher than the model suggests ($\mu > \$4000$). The observed sample mean of \$3050 is significantly lower than the theoretical mean of \$4000, leading to a very low Z-score that falls far outside the critical region for rejecting the null hypothesis in a one-tailed test with $\alpha = 0.05$.

# **CHI-SQUARE TEST**

# Association between Device Type and Customer Satisfaction

```python
import numpy as np
import scipy
from scipy import stats
import pandas as pd
```

```python
observed_data=pd.DataFrame({'Smart Thermostat': [50, 80, 60, 30, 20],
    'Smart Light': [70, 100, 90, 50, 50]
},
 index=['Very Satisfied', 'Satisfied', 'Neutral', 'Unsatisfied', 'Very Unsatisfied'])
```

```python
observed_data
```

**1.Stating the Hypotheses:**

```python
# H0: there's no significant association between the type of smart home device purchased and the customer satisfaction level.
# H1: there's a significant association between the type of smart home device purchased (Smart Thermostats vs. Smart Lights) and the customer satisfaction level.
```

**2.Computing the Chi-Square Statistic:**

```python
chi2_result = scipy.stats.chi2_contingency(observed_data)
print(chi2_result)
```

```python
statistic = chi2_result.statistic
print(statistic)
```

**3.Determine the Critical Value:**

```python
#significance level
alpha=0.05
```

```python
# To find the number of rows and columns
num_rows, num_cols = observed_data.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_cols}")
```

```python
#Degrees of freedom
df=(num_rows-1)*(num_cols-1)
print('Degrees of freedom: ',df)
```

```python
chi2_critic = stats.chi2.ppf(1 - alpha, df)
print('Chi-square critical value:', chi2_critic)
```

**4.Making a Decision:**

```python
if(statistic>chi2_critic):
  print('reject H0 and accept H1: there is a significant association between the type of smart home device purchased (Smart Thermostats vs. Smart Lights) and the customer satisfaction level.')
else:
  print('fail to reject H0: there is no significant association between the type of smart home device purchased and the customer satisfaction level.')
```

## Chi-Square Test Analysis Report

This report details the steps and findings of the Chi-Square test performed to investigate the association between device type (Smart Thermostat vs. Smart Light) and customer satisfaction level.

**1. Stating the Hypotheses:**

*   **Null Hypothesis ($H_0$):** There is no significant association between the type of smart home device purchased and the customer satisfaction level.
*   **Alternative Hypothesis ($H_1$):** There is a significant association between the type of smart home device purchased (Smart Thermostats vs. Smart Lights) and the customer satisfaction level.

**2. Computing the Chi-Square Statistic:**

The observed frequencies of customer satisfaction levels for each device type were provided in the `observed_data` DataFrame. The Chi-Square test statistic was calculated using `scipy.stats.chi2_contingency`.

The calculated Chi-Square test statistic is: **{{statistic}}**

**3. Determining the Critical Value:**

The significance level ($\alpha$) was set at 0.05. The degrees of freedom (df) were calculated based on the dimensions of the contingency table:

$df = (\text{Number of Rows} - 1) \times (\text{Number of Columns} - 1)$
$df = (5 - 1) \times (2 - 1) = 4 \times 1 = 4$

The critical value for the Chi-Square distribution with 4 degrees of freedom at a 0.05 significance level was determined using `scipy.stats.chi2.ppf`.

The Chi-Square critical value is: **{{chi2_critic}}**

**4. Making a Decision:**

To make a decision, the calculated Chi-Square test statistic is compared to the critical value.

*   If the test statistic > critical value, we reject the null hypothesis ($H_0$).
*   If the test statistic <= critical value, we fail to reject the null hypothesis ($H_0$).

In this case, the test statistic ({{statistic}}) is less than the critical value ({{chi2_critic}}).

Therefore, we **fail to reject the null hypothesis ($H_0$)**.

**5. Conclusion:**

Based on the Chi-Square test results, there is not enough statistical evidence at the 5% significance level to conclude that there is a significant association between the type of smart home device purchased and the customer satisfaction level. The observed differences in satisfaction levels between Smart Thermostats and Smart Lights could be due to random chance.
