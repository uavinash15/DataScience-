# **a. Building 99% Confidence Interval Using Sample Standard Deviation**

```python
import numpy as np
pheads=[1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
pheads=np.array(pheads)
pheads
```

```python
n=len(pheads)
print('sample size: ',n)
```

```python
pheads_mean=pheads.mean()
print('sample mean: ',pheads_mean)
```

```python
pheads_std=pheads.std()
print('sample standard deviation: ',pheads_std)
```

```python
df=n-1
print('degrees of freedom: ',df)
```

```python
CL=0.99
print('confidence level: ',CL)
```

```python
alpha=1-CL
print('significance level: ',alpha)
```

```python
from scipy import stats
tcritical_val=stats.t.ppf((CL+(alpha/2)),df)
print('t critical value: ',tcritical_val)
```

Calculating Margin of Error(E)

```python
std_error=((pheads_std)/(np.sqrt(n)))
print('Standared Error: ',std_error)
```

```python
E=tcritical_val*std_error
print('Margin of Error: ',E)
```

Finding Confidence Interval CI

```python
lower_bound=pheads_mean-E
upper_bound=pheads_mean+E
CI=(np.round(lower_bound,3),np.round(upper_bound,3))
print('Confidence Interval (CI): ',CI)
```

```python
# Explanation of steps and t-distribution rationale

# The first part of the notebook calculates a 99% confidence interval for the population mean using the sample standard deviation.
# Here's a breakdown of the steps:
# 1. Data Loading and Preparation: The sample data `pheads` was loaded and converted into a numpy array.
# 2. Sample Statistics Calculation: The sample size (n), sample mean, and sample standard deviation were calculated from the data.
# 3. Degrees of Freedom: The degrees of freedom (df) were calculated as n-1. This is used for the t-distribution.
# 4. Confidence and Significance Levels: The desired confidence level (CL) and the corresponding significance level (alpha) were defined.
# 5. Critical Value: The t-critical value was found using the `stats.t.ppf` function from the `scipy.stats` library. This function gives the inverse of the t-distribution's cumulative distribution function. We use `(CL + alpha/2)` to get the critical value for a two-tailed test.
# 6. Standard Error: The standard error of the mean was calculated using the sample standard deviation and the sample size.
# 7. Margin of Error: The margin of error (E) was calculated by multiplying the t-critical value and the standard error.
# 8. Confidence Interval: The lower and upper bounds of the confidence interval were calculated by subtracting and adding the margin of error from the sample mean, respectively.

# Rationale for using the t-distribution:
# We use the t-distribution because the population standard deviation is unknown and we are using the sample standard deviation as an estimate.
# The t-distribution is similar to the normal distribution but has heavier tails, which accounts for the increased uncertainty when using a sample standard deviation, especially with small sample sizes. As the sample size increases, the t-distribution approaches the normal distribution.
```

# **b. Building 99% Confidence Interval Using Known Population Standard Deviation**

finding point Estimate or mean

```python
pstd=0.2
print('population standard deviation: ',pstd)
```

Deciding confidence Levels

```python
CL=0.99
print('confidence Level: ',CL)
```

```python
alpha=1-CL
print('significance Level: ',alpha)
```

finding Margin of Error(E)

```python
z_critic_val=stats.norm.ppf(1-(alpha/2))
print('z critical value: ',z_critic_val)
```

```python
std_error=pstd/(np.sqrt(n))
print('standard Error: ',std_error)
```

```python
E=z_critic_val*std_error
print('Margin of Error: ',E)
```

finding confidence Interval(CI)

```python
lower_bound=pheads_mean-E
upper_bound=pheads_mean+E
CI=(np.round(lower_bound,3),np.round(upper_bound,3))
print('confidence Interval: ',CI)
```
