# **Exercise 1: Prime Numbers**

```python
import math

# Function to check if a number is prime
def is_prime(num):
  # Numbers less than or equal to 1 are not prime
  if num <= 1:
    return False
  # Check for divisibility from 2 up to the square root of the number
  for i in range(2, int(math.sqrt(num)) + 1):
    if num % i == 0:
      return False
  # If no divisors are found, the number is prime
  return True

try:
  # Get input from the user
  number = int(input("Enter a number for checking if it's prime: "))
  # Check if the number is prime and print the result
  if is_prime(number):
    print(f"{number} is a prime number")
  else:
    print(f"{number} is not a prime number")
except ValueError:
  # Handle cases where the input is not a valid integer
  print("please enter a valid integer")
```

# **Exercise 2: Product of Random Numbers**

```python
import random

# Generate two random integers between 1 and 100
number1 = random.randint(1, 100)
number2 = random.randint(1, 100)

# Ask the user for the product of the two numbers
answer = int(input(f"what's the value of {number1} * {number2} ? "))

# Check if the user's answer is correct and provide feedback
if answer == number1 * number2:
  print(f"your answer {answer} is correct")
else:
  print(f"your answer {answer} is wrong, correct answer is {number1 * number2}")
```

# **Exercise 3:Squares of Even/Odd Numbers**

```python
# Choosing even numbers because they are evenly divisible by 2 and simpler to process.
print('Squares of all Even numbers: ')
# Iterate through numbers from 100 to 200 (inclusive)
for num in range(100, 200 + 1):
  # Check if the number is even
  if num % 2 == 0:
    Even_num = num
    # Print the even number and its square
    print(f"Even Number: {Even_num}  square is : {Even_num**2}")
```

# **Exercise 4: Word counter**

```python
# Input text for word counting
input_text = "This is a sample text. This text will be used to demonstrate the word counter."
# Split the input text into words
words = input_text.split()
#print(words) # Uncomment to see the list of words

# Initialize an empty dictionary to store word counts
word_count_dict = {}
# Iterate through the list of words
for word in words:
  # If the word is already in the dictionary, increment its count
  if word in word_count_dict:
    word_count_dict[word] += 1
  # If the word is not in the dictionary, add it with a count of 1
  else:
    word_count_dict[word] = 1

# Print the word counts
for word, count in word_count_dict.items():
  print(f"'{word}': {count}")
```

# **Exercise 5: Check for Palindrome**

```python
# Function to check if a string is a palindrome
def is_palindrome(string):
  # Check if the string is equal to its reverse
  if string == string[::-1]: # Use slicing to reverse the string
    return True
  else:
    return False

# Get input text from the user
input_text = input('Enter the text that you want to check for palindrome: ')
# Check if the input text is a palindrome and print the result
if is_palindrome(input_text):
  print(True)
else:
  print(False)
```
