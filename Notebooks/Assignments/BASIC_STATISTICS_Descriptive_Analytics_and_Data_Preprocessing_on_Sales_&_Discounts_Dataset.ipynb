# **Standardization of Numerical Variables**

z score( standardization)
z=(x-Mean)/std
z - z score.
x - each value,
mean - mean of the data.
std- standard deviation
std= sqrt(variance)



**Standardization (Z-score Normalization)**

Standardization, also known as Z-score normalization, is a data preprocessing technique used to transform numerical data so that it has a mean of 0 and a standard deviation of 1. This is achieved by subtracting the mean of the data from each data point and then dividing by the standard deviation.

The formula for calculating the Z-score for a data point 'x' is:

$z = (x - \mu) / \sigma$

Where:
- $z$ is the Z-score of the data point.
- $x$ is the individual data point.
- $\mu$ is the mean of the data.
- $\sigma$ is the standard deviation of the data.

Standardization is particularly useful when working with machine learning algorithms that are sensitive to the scale of the input features. By standardizing the data, you ensure that all features contribute equally to the model's training process, preventing features with larger values from dominating those with smaller values. This can lead to improved model performance and faster convergence during training.

In this notebook, we will apply standardization to the numerical columns of the sales dataset to prepare it for further analysis or modeling.

```python
#volume mean,std,zscore
mean_volume = sales_numeric_df['Volume'].mean()
std_volume = sales_numeric_df['Volume'].std()
sales_numeric_df['Volume_zscore'] = (sales_numeric_df['Volume'] - mean_volume) / std_volume
sales_numeric_df[['Volume','Volume_zscore']].head()
```

```python
mean_price = sales_numeric_df['Avg Price'].mean()
std_price = sales_numeric_df['Avg Price'].std()
sales_numeric_df['Avg Price_zscore'] = (sales_numeric_df['Avg Price'] - mean_price) / std_price
display(sales_numeric_df[['Avg Price', 'Avg Price_zscore']].head())
```

```python
mean_total_sales = sales_numeric_df['Total Sales Value'].mean()
std_total_sales = sales_numeric_df['Total Sales Value'].std()
sales_numeric_df['Total Sales Value_zscore'] = (sales_numeric_df['Total Sales Value'] - mean_total_sales) / std_total_sales
display(sales_numeric_df[['Total Sales Value', 'Total Sales Value_zscore']].head())
```

```python
mean_net_sales = sales_numeric_df['Net Sales Value'].mean()
std_net_sales = sales_numeric_df['Net Sales Value'].std()
sales_numeric_df['Net Sales Value_zscore'] = (sales_numeric_df['Net Sales Value'] - mean_net_sales) / std_net_sales
display(sales_numeric_df[['Net Sales Value', 'Net Sales Value_zscore']].head())
```

```python
mean_discount_amount = sales_numeric_df['Discount Amount'].mean()
std_discount_amount = sales_numeric_df['Discount Amount'].std()
sales_numeric_df['Discount Amount_zscore'] = (sales_numeric_df['Discount Amount'] - mean_discount_amount) / std_discount_amount
display(sales_numeric_df[['Discount Amount', 'Discount Amount_zscore']].head())
```

```python
mean_discount_rate = sales_numeric_df['Discount Rate (%)'].mean()
std_discount_rate = sales_numeric_df['Discount Rate (%)'].std()
sales_numeric_df['Discount Rate (%)_zscore'] = (sales_numeric_df['Discount Rate (%)'] - mean_discount_rate) / std_discount_rate
display(sales_numeric_df[['Discount Rate (%)', 'Discount Rate (%)_zscore']].head())
```

**Importance of Data Preprocessing: Standardization and One-Hot Encoding**

The data preprocessing steps performed in this notebook, specifically standardization (Z-score normalization) and one-hot encoding (dummy variables), are crucial for preparing the data for further analysis and machine learning model building.

* **Standardization:** As seen in the descriptive statistics and visualizations, several numerical columns exhibit significant skewness and contain outliers. Standardization helps to transform these variables so that they have a mean of 0 and a standard deviation of 1. This is particularly important for algorithms that are sensitive to the scale of features, such as:
    * **Distance-based algorithms:** K-Nearest Neighbors (KNN), Support Vector Machines (SVM), K-Means Clustering. These algorithms calculate distances between data points, and features with larger scales can disproportionately influence these distances.
    * **Gradient descent-based algorithms:** Linear Regression, Logistic Regression, Neural Networks. Standardization can help these algorithms converge faster and avoid getting stuck in local minima.

* **One-Hot Encoding:** The categorical columns in the dataset contain nominal data (categories without an inherent order). Machine learning algorithms typically require numerical input. One-hot encoding converts these categorical variables into a numerical format that can be understood by these algorithms. By creating dummy variables for each category, we avoid implying any ordinal relationship between categories and prevent the model from misinterpreting categorical data as having a numerical order.

**In summary,** standardization and one-hot encoding are essential preprocessing steps that address issues related to feature scaling, skewed distributions, and the representation of categorical data. By applying these techniques, we ensure that the data is in a suitable format for various analytical techniques and machine learning algorithms, leading to more accurate and reliable results.
