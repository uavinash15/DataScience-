```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 1, 4])
markers = ['o', 's', '^', 'D', '*']
#print(range(len(x)))
for i in range(len(x)):
    plt.plot(x[i], y[i], marker=markers[i], markersize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 1, 4])
markers = ['o', 's', '^', 'D', '*']
colors = ['red', 'blue', 'green', 'orange', 'purple']
labels = ['Circle', 'Square', 'Triangle', 'Diamond', 'Star']

for i in range(len(x)):
    plt.plot(x[i], y[i], marker=markers[i], color=colors[i],
             markersize=10, label=labels[i], linestyle='')  # No line between points

plt.legend()
plt.title("Different Matplotlib Markers")
plt.grid(True)
plt.show()
```
