# importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
X=np.random.uniform(0,1,1000)
Y=np.random.uniform(0,1,1000)
print(X)
```

```python
print(Y)
```

```python
df_xy=pd.DataFrame(columns=['X','Y'])
df_xy
df_xy.X=X
df_xy
df_xy.Y=Y
df_xy
df_xy.plot(x='X',y='Y',kind='scatter')
model1=KMeans(n_clusters=5).fit(df_xy)
plt.show()
```

```python
#model1.labels_
```

```python
df_xy.plot(x='X',y='Y',c=model1.labels_,kind='scatter',s=50,cmap=plt.cm.coolwarm_r)
plt.show()
```

```python
Univ=pd.read_csv('/content/Universities_Clustering.csv')
Univ
```

```python
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)
```

```python
df_norm=norm_func(Univ.iloc[:,1:])
```

```python
df_norm
```

```python
from sklearn.cluster import KMeans
fig=plt.figure(figsize=(10,8))
WCSS=[]
for i in range(1,11):
  #print(i)
  clf=KMeans(n_clusters=i) #classifier
  #print(clf)
  clf.fit(df_norm)
  #print(clf)
  #print('inertia: ',clf.inertia_)
  WCSS.append(clf.inertia_)
plt.plot(range(1,11),WCSS)
plt.title('The Elbow Method Curve')
plt.xlabel('no of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clf=KMeans(n_clusters=5,random_state=42)
y_kmeans=clf.fit(df_norm)
```

```python
y_kmeans
```

```python
clf.cluster_centers_ #gives 5 values rows, and 6 columns because there are 6 columns we selected in university_clustering.csv file
```

```python
clf.labels_
```

```python
clf.inertia_ #gives at k=5 WCSS(within clusters sum of square)
```
