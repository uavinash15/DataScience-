# **K-MEANS CLUSTERING**

```python
from sklearn.cluster import KMeans
clusters=KMeans(3)
clusters.fit(customers_df)
```

```python
clusters.labels_
```

```python
customers_df["clusterid"]=clusters.labels_
```

```python
customers_df.head()
```

# Plotting Customers with Their Segments

```python
markers=['+','^','.']
sn.lmplot(x="age",y="income",data=customers_df,hue="clusterid",fit_reg=False,markers=markers)  #lmplot=linearmodel plot
plt.show()
```
