```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
Univ=pd.read_csv('/content/Universities_Clustering.csv')
Univ
```

# standardize data: z score transform

```python
# x'=(x-xmin)/(xmax-xmin)
```

```python
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)
```

```python
df_norm=norm_func(Univ.iloc[:,1:])
df_norm
```

```python
from sklearn.preprocessing import MinMaxScaler
trans=MinMaxScaler()
data=pd.DataFrame(trans.fit_transform(Univ.iloc[:,1:]))
data
```

```python

from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z=linkage(df_norm,method="average",metric="euclidean")
plt.figure(figsize=(15,5))
plt.title('Hierarchial Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
h_complete=AgglomerativeClustering(n_clusters=5,linkage='average').fit(df_norm)
#h_complete.labels_
cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
Univ['clust']=cluster_labels
Univ
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
Univ_subset=pd.DataFrame(scaler.fit_transform(Univ.iloc[:,1:7]))
Univ_subset
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z=linkage(Univ_subset,method="average",metric="euclidean")
plt.figure(figsize=(15,5))
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Index")
plt.ylabel("Distance")
labels = [f"{univ}  {i}" for i, univ in enumerate(Univ['Univ'].tolist())]
sch.dendrogram(z, labels=labels
               , leaf_rotation=80
)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
h_complete=AgglomerativeClustering(n_clusters=5,linkage='average').fit(Univ_subset)
cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
Univ['clust']=cluster_labels
Univ
```
