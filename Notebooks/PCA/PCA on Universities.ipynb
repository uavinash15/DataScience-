```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
uni = pd.read_csv("Universities_Clustering (1).csv")
uni
```

```python
uni.describe()
```

```python
uni.head()
```

```python
# Considering only numerical data
uni.data = uni.iloc[:,1:]
uni.data.head()
# Converting into numpy array
UNI = uni.data.values
UNI
```

```python
# Normalizing the numerical data
uni_normal = scale(UNI)
```

```python
uni_normal
```

```python
pca = PCA(n_components = 6)
pca_values = pca.fit_transform(uni_normal)
pca_values
```

```python
#loadings or weights
pca.components_
```

```python
# The amount of variance that each PCA explains is
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
# Variance plot for PCA components obtained
plt.plot(var1,color="red")
plt.show()
```

```python
pca_values[:,0:2]
```

```python
pd.concat([pd.DataFrame(pca_values[:,0:2],columns=['pc1','pc2']), uni[['Univ']]],axis=1)
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values[:,0:2],columns=['pc1','pc2']), uni[['Univ']]], axis = 1)
finalDf
```

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
import seaborn as sns
sns.scatterplot(data=finalDf,x='pc1',y='pc2',hue='Univ',s = 100)
plt.show()
```
